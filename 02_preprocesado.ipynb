## Drop the repeated column **FAMI_TIENEINTERNET.1**

```python
train_df = train_df.drop(columns="FAMI_TIENEINTERNET.1")
train_df
```

```python
num_variables_unicas = train_df['ESTU_PRGM_ACADEMICO'].nunique()
print("El número de variables únicas en la columna ESTU_PRGM_ACADEMICO es:", num_variables_unicas)
```

```python
# Total de valores NaN en el DataFrame
total_nans = train_df.isna().sum().sum()
total_datos = train_df.size

# Porcentaje de valores NaN en el total de datos
porcentaje_nans = (total_nans / total_datos) * 100


print("El DataFrame tiene un total de {} datos.".format(total_datos))
print("El DataFrame tiene un total de {} valores NaN.".format(total_nans))
print("El {:.2f}% de los datos en el DataFrame son valores NaN.".format(porcentaje_nans))

columna_con_mas_nan = train_df.isna().sum().idxmax()
cantidad_nan = train_df.isna().sum().max()

```

# Fill null data from ESTU_VALORMATRICULAUNIVERSIDAD with the mean

```python
train_df['ESTU_VALORMATRICULAUNIVERSIDAD'] = train_df['ESTU_VALORMATRICULAUNIVERSIDAD'].fillna(train_df['ESTU_VALORMATRICULAUNIVERSIDAD'].mode().iloc[0])

train_df
```

# fill null data from ESTU_HORASSEMANATRABAJA with 0

```python
train_df['ESTU_HORASSEMANATRABAJA'] = train_df['ESTU_HORASSEMANATRABAJA'].fillna(0)
train_df
```

# Replace letters with accent with letters without accent in the column ESTU_PRGM_ACADEMICO

```python
def reemplazar_letra(palabra,letra,letraSinAcento):
  return palabra.replace(letra,letraSinAcento)

train_df['ESTU_PRGM_ACADEMICO'] = train_df['ESTU_PRGM_ACADEMICO'].apply(lambda x: reemplazar_letra(x,'Á','A'))
train_df['ESTU_PRGM_ACADEMICO'] = train_df['ESTU_PRGM_ACADEMICO'].apply(lambda x: reemplazar_letra(x,'É','E'))
train_df['ESTU_PRGM_ACADEMICO'] = train_df['ESTU_PRGM_ACADEMICO'].apply(lambda x: reemplazar_letra(x,'Í','I'))
train_df['ESTU_PRGM_ACADEMICO'] = train_df['ESTU_PRGM_ACADEMICO'].apply(lambda x: reemplazar_letra(x,'Ó','O'))
train_df['ESTU_PRGM_ACADEMICO'] = train_df['ESTU_PRGM_ACADEMICO'].apply(lambda x: reemplazar_letra(x,'Ú','U'))

train_df
```

# Fill Na data in FAMI_TIENEINTERNET with no

```python
train_df['FAMI_TIENEINTERNET'] = train_df['FAMI_TIENEINTERNET'].fillna('No')
train_df
```

# Fill Na data in FAMI_EDUCACIONPADRE	with No Sabe

```python
train_df['FAMI_EDUCACIONPADRE'] = train_df['FAMI_EDUCACIONPADRE'].fillna('No sabe')
train_df
```

# Fill Na data in ESTU_PAGOMATRICULAPROPIO with No

```python
train_df['ESTU_PAGOMATRICULAPROPIO'] = train_df['ESTU_PAGOMATRICULAPROPIO'].fillna('No')
train_df
```

# Fill Na data in FAMI_ESTRATOVIVIENDA with No responde

```python
train_df['FAMI_ESTRATOVIVIENDA'] = train_df['FAMI_ESTRATOVIVIENDA'].fillna('No responde')

train_df
```

```python
# Total de datos en el DataFrame
total_datos = train_df.size

# Total de valores NaN en el DataFrame
total_nans = train_df.isna().sum().sum()

# Porcentaje de valores NaN en el total de datos
porcentaje_nans = (total_nans / total_datos) * 100

print("tiene un total de {} datos.".format(total_datos))
print(" total de {} valores NaN.".format(total_nans))
print("El {:.2f}% de los datos en el DataFrame son valores NaN.".format(porcentaje_nans))

columna_con_mas_nan = train_df.isna().sum().idxmax()
cantidad_nan = train_df.isna().sum().max()

print("La columna con el mayor número de valores NaN es '{}' con {} valores NaN.".format(columna_con_mas_nan, cantidad_nan))
```

```python
valores_nan_por_columna = train_df.isna().sum()
print(valores_nan_por_columna)
```

```python
#correcciones especificas para los NaN faltantes
train_df['FAMI_TIENELAVADORA'] = train_df['FAMI_TIENELAVADORA'].fillna('No')
train_df['FAMI_TIENEAUTOMOVIL'] = train_df['FAMI_TIENEAUTOMOVIL'].fillna('No')
train_df['FAMI_TIENECOMPUTADOR'] = train_df['FAMI_TIENECOMPUTADOR'].fillna('No')
train_df['FAMI_EDUCACIONMADRE'] = train_df['FAMI_EDUCACIONMADRE'].fillna('No sabe')
```

```python
valores_nan_por_columna = train_df.isna().sum()
print(valores_nan_por_columna)
```

## One-hot encoding the corresponding columns

```python
yes_no_columns = ["FAMI_TIENEINTERNET", "FAMI_TIENELAVADORA",'FAMI_TIENEAUTOMOVIL', 'ESTU_PRIVADO_LIBERTAD',
       'ESTU_PAGOMATRICULAPROPIO', 'FAMI_TIENECOMPUTADOR','ESTU_HORASSEMANATRABAJA','ESTU_PRGM_DEPARTAMENTO','ESTU_VALORMATRICULAUNIVERSIDAD']

train_df = pd.get_dummies(train_df, columns=yes_no_columns)
```

```python

train_df
```

```python
num_variables_unicas = train_df['ESTU_PRGM_ACADEMICO'].nunique()
print("El número de variables únicas en la columna ESTU_PRGM_ACADEMICO es:", num_variables_unicas)
```

```python
#eliminamos espacios en blanco
train_df['ESTU_PRGM_ACADEMICO'] = train_df['ESTU_PRGM_ACADEMICO'].str.strip().str.lower()
```

```python
num_variables_unicas = train_df['ESTU_PRGM_ACADEMICO'].nunique()
print("El número de variables únicas en la columna ESTU_PRGM_ACADEMICO es:", num_variables_unicas)
```

```python
# Contar las variables únicas que contienen el signo "?"
num_variables_con_signo = train_df['ESTU_PRGM_ACADEMICO'].str.contains('\¿').sum()

print("Número de variables únicas que contienen el signo '¿':", num_variables_con_signo)
```

```python
import re
#. La biblioteca re permite realizar diversas operaciones relacionadas con expresiones regulares, como
#buscar, extraer,reemplazar y dividir cadenas basadas en patrones específicos.

# Concatenar todas las cadenas en una sola y eliminar las letras
texto_concatenado = ''.join(train_df['ESTU_PRGM_ACADEMICO'])
caracteres_no_letras = re.sub(r'[a-zA-Z\s]', '', texto_concatenado)

# Obtener caracteres únicos aparte de las letras
caracteres_unicos = set(caracteres_no_letras)

print("Caracteres aparte de letras que aparecen en la columna ESTU_PRGM_ACADEMICO:", caracteres_unicos)
```
